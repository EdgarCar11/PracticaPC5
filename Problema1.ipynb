```python
# Filtrar con precio de 50€ o menores y de tipo Shared room
df_diana = df_airbnb[(df_airbnb['price'] <= 50) & (df_airbnb['room_type'] == 'Shared room')]

# Ordenarde mayor puntuacion a menor puntuacion
df_diana_sorted = df_diana.sort_values(by='overall_satisfaction', ascending=False)

# Las 10 propiedades 
df_diana_sorted.head(10)
```

```python
#precio promedio por barrio
df_barrio_precio = df_airbnb.groupby('neighborhood')['price'].mean()

# resultado
df_barrio_precio
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico de precio promedio por barrio
plt.figure(figsize=(10, 6))
sns.barplot(x=df_barrio_precio.index, y=df_barrio_precio.values)
plt.xticks(rotation=90)
plt.title('Precio promedio por barrio')
plt.xlabel('Barrio')
plt.ylabel('Precio promedio (€)')
plt.show()
```

```python
# Agrupar por tipo de propiedad y calcular la satisfacción promedio
df_tipo_satisfaccion = df_airbnb.groupby('room_type')['overall_satisfaction'].mean()

# resultado
df_tipo_satisfaccion
```

```python
# Gráfico de satisfacción promedio por tipo de propiedad
plt.figure(figsize=(8, 5))
sns.barplot(x=df_tipo_satisfaccion.index, y=df_tipo_satisfaccion.values)
plt.title('Satisfacción promedio por tipo de propiedad')
plt.xlabel('Tipo de propiedad')
plt.ylabel('Satisfacción promedio')
plt.show()
```
